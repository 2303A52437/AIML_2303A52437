### 1. Implement simple linear regression on a dataset and plot the regression line.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('/content/hw_200 (3).csv')
data.head()
```

```python
data.columns
```

### Simple Linear Regression

```python
x = data[' Height(Inches)"']
y = data[' "Weight(Pounds)"']
```

```python
model = LinearRegression()
model.fit(x.values.reshape(-1, 1), y)
```

```python
y_pred = model.predict(x.values.reshape(-1, 1))
```

# Plot

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y, color = 'blue', label = 'Actual Data')
plt.plot(x, y_pred, color = 'red', label = 'Predicted Line')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```

### 2. Calculate the R-squared value of the model and interpret it.

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
ss_tot = np.sum((y-np.mean(y)**2))
ss_res = np.sum((y-y_pred)**2)
r2 = 1 - (ss_res/ss_tot)
print(r2)
```

### 3. Perform multiple linear regression with 2-3 independent variables and compare the results.

```python
X = data[' Height(Inches)"']
Y = data[' "Weight(Pounds)"']
```

```python
#split the data int train and test

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train.values.reshape(-1, 1), Y_train)
```

```python
model_pred = model.predict(X_test.values.reshape(-1, 1))
```

```python
mse = mean_squared_error(Y_test, model_pred)
r2 = r2_score(Y_test, model_pred)
print("R-squared:", r2)
print("Mean Squared Error:", mse)
```

### 4. Plot the residuals and check for any patterns.

```python
residuals = Y_test - model_pred

## Plot residuals
plt.figure(figsize=(10,6))
plt.scatter(model_pred, residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
```
